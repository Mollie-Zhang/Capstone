```python
#2022 Yearly Electricity Genertation
# List of the 12 DataFrames for each month
dataframes = [elec_01, elec_02, elec_03, elec_04, elec_05, elec_06, elec_07, elec_08, elec_09, elec_10, elec_11, elec_12]

# Concatenate the DataFrames vertically to get the yearly electricity generation dataframe
e2022_df = pd.concat(dataframes, ignore_index=True)

e2022_df
```

```python
#drop the 4th to 15th columns
e2022_df = e2022_df.drop(e2022_df.columns[4:15], axis=1)
e2022_df
```

```python
# 2022 yearly elecrtricity generation aggragation by 'Month' and'DUID' 
# Group by 'Month' and 'DUID', and sum the 'SCADAVALUE' - i.e. the capacity of generator
aggregated_df = e2022_df.groupby(['Month', 'DUID'])['SCADAVALUE'].sum().reset_index()
aggregated_df
```

```python
aggregated_df.isnull().sum()
```

```python
#Save the aggregated sheet as the yearly aggregated dataset
aggregated_df.to_csv('aggregatedDUID_2022.csv', index=False)
```

```python
#Open the dataset of the generator information - 'Generators.csv'

file_generators = 'Generators.csv'

generators = pd.read_csv(file_generators)

generators
```

```python
#Drop'Unnamed' columns with null values
generators.drop(generators.columns[6:8], axis=1, inplace=True)
generators
```

```python
#Merge aggregated_df and generators using LEFT JOIN

merged_df = aggregated_df.merge(generators, on='DUID', how='left')
merged_df
```

```python
merged_df.isnull().sum()
```

```python
merged_df.head(20)
```

```python
#Save the aggregated sheet as the yearly aggregated dataset
merged_df.to_csv('merged_2022.csv', index=False)
```

```python
#Drop the rows of which the value in the 'SCADAVALUE' equals or is less than 0
#Drop the rows of which the value in the 'Dispatch Type' equals 'Load' - i.e. they are not generators

temp_df = merged_df[(merged_df['SCADAVALUE'] > 0) & (merged_df['Dispatch Type'] !='Load')]
temp_df
```

```python
#Add a new column 'Electricity Generation(MWh)'
temp_df['Electricity Generation (MWh)'] = round(temp_df['SCADAVALUE'] * (5/60), 2)
temp_df
```

```python
temp_df.to_csv('temp_2022.csv', index=False)
```
